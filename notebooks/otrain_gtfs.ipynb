```python
# Loading of otrain data. Copied from Open Train notebook

# library for loading data and tables
# we will use it a lot, so we use its standard abbreviation - pd
import pandas as pd
import numpy as np

# plotting library
# loading it changes default styles to much nicer ones
import seaborn as sns

# IPython Notebook option to show plots in the notebook (not in a separate window)
%matplotlib inline

```

```python
df = pd.read_excel("data/otrain/2015_04_05_06.xlsx")
```

```python
df['Planned_Stop'].fillna(0, inplace=True)
df['Actual_Stop'].fillna(0, inplace=True)
di = {'מתוכננת': True, 'לא מתוכננת': False}
df.replace({"Planned": di}, inplace=True)
df = df.rename(columns={'Station_typ': "Station_type", "Planned_Depratur": "Planned_Deprature"})
df = df[(df["Station_type"] != 'מוצא תפעולי') & (df["Station_type"] != "יעד תפעולי")]
df.ix[df["Station_type"] == "יעד מסחרי", "Actual_Departure"] = pd.NaT
df.ix[df["Station_type"] == "מוצא מסחרי", "Actual_Arrival"] = pd.NaT

```

```python
df_full = df
df = df_full[df_full.Train_Date == '2015-04-01']
```

```python
# add route id field to make things easier 
routes_heb = ['נהריה - נתבג', 
             'נהריה - באר שבע מרכז',
             'נהריה - מודיעין מרכז',
             'בנימינה - אשקלון',
             'נהריה - חוף הכרמל',
             'הרצליה- ירושלים מלחה',
             'הוד השרון-שדרות',
             'חיפה מרכז - בש מרכז',
             'באר שבע צפון-דימונה',
             'נתניה-הראשונים',
             'קו לא מוגדר', 
             np.nan]

routes_eng = ['Nahariyya - Ben Gurion Airport',
              "Nahariyya - Be'er Sheva Center",
              'Nahariyya - Modiin Center',
              'Binyamina - Ashkelon',
             "Nahariyya - Haifa Hof HaKar     mel (Razi'el)",
             'Hertsliya - Jerusalem Malha',
             'Hod HaSharon - Sderot',
             "Haifa Center-Be'er Sheva Center",
             "Be'er Sheva North University - Dimona",
             'Netanya - Rishon LeTsiyyon HaRishonim',
             'undefined route',
             np.nan]

heb_to_eng = {x: y for x,y in zip(routes_heb, routes_eng)}

from csv import DictReader
from io import TextIOWrapper
f = open('data/gtfs/routes.txt', encoding='utf-8-sig')
name_to_id = {row['route_long_name']: row['route_id'] for row in  DictReader(f)}
name_to_id[np.nan] = np.nan
name_to_id['undefined route'] = '0'
heb_to_id = {x: name_to_id[heb_to_eng[x]] for x in routes_heb}
df['Route_Id'] = df['Route_Description'].map(heb_to_id)
df.head()
```
